```python
import pandas as pd

data = {
    'Date': ['2025-01-01', '2025-01-01', '2025-01-02', '2025-02-01', '2025-02-02'],
    'Product': ['Laptop', 'Mobile', 'Laptop', 'Tablet', 'Mobile'],
    'Sales': [1500, 800, 1300, 900, 700]
}

df = pd.DataFrame(data)
df.to_csv('sample_sales.csv', index=False)

# Load it
df = pd.read_csv('sample_sales.csv')
df.head()
```

```python
# Overview of columns and data types
df.info()

# Summary statistics
df.describe()

# Check for missing values
df.isnull().sum()

```

```python
sales_by_product = df.groupby('Product')['Sales'].sum().sort_values(ascending=False)
print(sales_by_product)

# Bar plot
sales_by_product.plot(kind='bar', figsize=(10, 5), title='Total Sales by Product')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Convert to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract month
df['Month'] = df['Date'].dt.to_period('M')

# Group by Month
monthly_sales = df.groupby('Month')['Sales'].sum()

# Line plot
monthly_sales.plot(kind='line', marker='o', figsize=(12, 6), title='Monthly Sales Trend')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
